```python
import pandas as pd
import altair as alt 
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('revenue file.csv')
df
```

```python
df
```

```python
total = df['TOTAL'].str.replace(',', '').astype(float)
```

```python
x=df['TEAMS']
y=total
```

```python
df
```

```python
plt.bar(x,y,color='purple')
plt.title('THIS IS A SCATTERPLOT SHOWING REVENUE COLLECTION')
plt.xlabel('TEAMS')
plt.ylabel('TOTAL')
plt.show()
```
